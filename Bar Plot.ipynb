```python
'''
Plot the bar for students of 3 classes join lec for Python ,R , AI , ML , DS
'''
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
classes = ['python','R','AI','ML','DS']
class_1 = [30,10,20,25,10] 
class_2 = [40,5,20,20,10]
class_3 = [35,5,30,15,15]

class_index = np.arange(len(classes))
width=0.2

plt.figure(figsize=(10,10))

plt.bar(class_index , class_1,width, color='g',edgecolor = 'm' , linestyle='-.',
       label='Students of Class 1')

plt.bar(class_index + width, class_2,width , color='b',edgecolor = 'k', linestyle='-.',
       label='Students of Class 2')

plt.bar(class_index + 2*width, class_3,width , color='c',edgecolor = 'y', linestyle='-.',
       label='Students of Class 3')

plt.legend()
plt.xticks(class_index + width , classes,rotation=45)
plt.title("Bar chart of IAIP class",size=30)
plt.xlabel("Classes",size=20)
plt.ylabel("No. of Students",size=20)
plt.show()
```
