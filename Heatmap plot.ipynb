```python
'''
Plotting Heatmap of countries contribution to Global Warming
'''
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
globalwarming_df = pd.read_csv("F:\\Data\\Who_is_responsible_for_global_warming.csv")
globalwarming_df.head()
```

```python
globalwarming_df.corr().head()
```

```python
plt.figure(figsize=(16,9))
ax = sns.heatmap(globalwarming_df.corr() , annot=True , linewidth=3)

ax.tick_params(size=10,color='w',labelsize=10,labelcolor='k')
plt.title("Who is respomsible for 'Global Warming'")
plt.show()
```
